## Advance operation with business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure:[restaurant_id, 2021,2022,2023,2024]

sales_data = np.array([
    [1,150000,180000,220000,250000],
    [2,120000,140000,160000,190000],
    [3,200000,230000,260000,300000],
    [4,180000,210000,240000,270000],
    [5,160000,185000,205000,230000],
])

print("=== Zomato Sales Analysis === ")
print("\n Sales data shape ", sales_data.shape)
print("\n Sales data size ",sales_data.size)
print("\n Sales data dimensions ",sales_data.ndim)
print("\n Sales data datatype ",sales_data.dtype)
print("\n Sales data from first three restaurants \n",sales_data[0:3])
print("\n Sales data from all the rows \n",sales_data[:, 1:])
```

```python
# 0 is for top to bottom ops(columns) and 1 is for left to right ops(rows)
# total sales per year
print(np.sum(sales_data, axis= 0))

# removing the index column for calculating yearly total
print(np.sum(sales_data[:,1:], axis= 0))

# minimum sales per restaurant
min_sales = np.min(sales_data[:,1:],axis=1)
print(min_sales)

# maximum sales per year
max_sales = np.max(sales_data[:,1:],axis=0)
print(max_sales)

# average sales per restaurant
avg_sales = np.mean(sales_data[:,1:],axis=1)
print(avg_sales)

# cumulative sales
cumsum = np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)
```

```python
plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("average cumulative sales across all restaurant")
plt.xlabel("years")
plt.ylabel("sales")
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("vector addition: ",vector1+vector2)
```

```python
print("vector multiplication: ",vector1*vector2) # vector multiplication is also called as hadamard product which is used when doing element-wise multiplication. it requires input vectors of same size and the output is also the vector of same size 
```

```python
print("dot product: ",np.dot(vector1,vector2)) # aka inner product, the output will always be a scalar. multiply the existing vectors and get the sum of their product
```

```python
angle = np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print(angle)
```

```python
# vector ops
restaurant_types = np.array(['biryani','chinese','pizza','burger'])
vectorized_upper = np.vectorize(str.upper)
print("vectorized upper",vectorized_upper(restaurant_types))
```

```python
# broadcasting : performing ops on every element of the array
monthly_avg = sales_data[:,1:]/12
print(monthly_avg)
```
